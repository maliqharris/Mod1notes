## Assignment 1: Greeting and String Manipulation

**Goal:** Create a program that asks the user for their name and then prints a personalized greeting. The program should also show them the length of their name and display their name in uppercase letters.

**Step-by-Step Instructions:**  
1. Prompt the user to input their name using the `input()` function.  
   ```python
   name = input("Enter your name: ")
   ```
2. Print a greeting message that includes their name, for example:  
   ```python
   print("Hello, " + name + "! Nice to meet you.")
   ```
3. Find out how long the name is by using `len(name)` and print the length:  
   ```python
   length_of_name = len(name)
   print("Your name has " + str(length_of_name) + " characters.")
   ```
4. Print the name in uppercase by using `name.upper()` and show it to the user:  
   ```python
   print("In uppercase, your name looks like: " + name.upper())
   ```

**Things to Practice:**  
- Using `input()` to get user input.  
- Using `print()` to display messages.  
- Converting data types (e.g., `str(length_of_name)` to print numbers).  
- Calling string methods like `.upper()` and using `len()` to find the length.

## Assignment 2: Summation of Numbers

**Goal:** Create a program that asks the user for a positive number `n` and then prints the sum of all the numbers from 1 up to `n`. For example, if `n` = 5, the program should print the sum 1+2+3+4+5 = 15.

**Step-by-Step Instructions:**  
1. Prompt the user to input a positive integer:  
   ```python
   n = int(input("Enter a positive integer: "))
   ```
2. Initialize a variable `total` to 0. This will hold the running sum:  
   ```python
   total = 0
   ```
3. Use a `for` loop or `while` loop to add each number from 1 to `n` to the total:  
   ```python
   for i in range(1, n+1):
       total += i
   ```
4. Print the result:  
   ```python
   print("The sum of numbers from 1 to " + str(n) + " is " + str(total))
   ```

**Things to Practice:**  
- Using `int()` to convert input to an integer.  
- Using loops (`for` loops) and the `range()` function.  
- Basic arithmetic operations and printing the result.

## Assignment 3: Counting Vowels in a String

**Goal:** Create a program that asks the user for a sentence (or any string of text) and then counts how many vowels (a, e, i, o, u) are in that sentence. Print the total number of vowels found.

**Step-by-Step Instructions:**  
1. Prompt the user to enter a sentence:  
   ```python
   sentence = input("Enter a sentence: ")
   ```
2. Convert the sentence to lowercase to make checking easier (so that 'A' and 'a' are treated the same):  
   ```python
   sentence = sentence.lower()
   ```
3. Initialize a counter variable to 0 for counting vowels:  
   ```python
   vowel_count = 0
   ```
4. Loop through each character in the sentence. For each character, check if it is a vowel (a, e, i, o, u). If it is, increase the counter by 1:  
   ```python
   for char in sentence:
       if char in "aeiou":
           vowel_count += 1
   ```
5. Print the total number of vowels:  
   ```python
   print("The total number of vowels in your sentence is:", vowel_count)
   ```

**Things to Practice:**  
- String iteration (looping through each character in a string).  
- Using conditionals (`if`) to check if a character is a vowel.  
- Summarizing and printing out a final count.
